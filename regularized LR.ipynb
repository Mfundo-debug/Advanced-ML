```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate data
n = 100
x = np.random.uniform(0, 30, n)
y = 2 + 3 * x + 2 * x **2
y_o = y + y + np.random.normal(0, 90, n)
```

```python
#plot the points
plt.plot(x, y_o, '.')
plt.show()
```

```python
#create matrix design
X = np.c_[np.ones(n), x, x**2]
```

```python
#generate the parameters using LSR
θ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_o)
```

```python
θ
```

```python
#sort x values
xc = x[np.argsort(x)]
plt.plot(x, y_o, '.')
plt.plot(xc, θ[0] + θ[1]*xc + θ[2]*xc**2)
plt.show()
```

```python
 λ = 10
I = np.eye(3)
I[0,0] = 0
```

```python
θ_r = np.linalg.inv(X.T.dot(X) + λ * I).dot(X.T).dot(y_o)
θ_r
```

```python
xc = x[np.argsort(x)]
plt.plot(x, y_o, '.')
plt.plot(xc, θ_r[0] + θ_r[1]*xc + θ_r[2]*xc**2)
plt.show()
```
