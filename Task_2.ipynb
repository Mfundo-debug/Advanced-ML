```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 200
x = np.random.uniform(30,0,n)
```

```python
θ_0 = 1
θ_1 = 2
y = θ_0 + θ_1 * x
ϵ = np.random.normal(0,15,n)
y_obv = y + ϵ
```

```python
plt.plot(x,y_obv,'.')
```

```python
X = np.c_[np.ones(n),x]
```

```python
θ = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_obv)
```

```python
σ = (1/n)*((y_obv - X.dot(θ)).T.dot(y_obv - X.dot(θ)))
σ
```

```python
y_ = np.random.normal(X.dot(θ),σ)
y_
```

```python
xc = x[np.argsort(x)]
plt.plot(x, y_obv, '.')
plt.plot(xc, θ[0] + θ[1]*xc )
plt.show()
```
